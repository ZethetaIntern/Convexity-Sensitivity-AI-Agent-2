```python
import pandas as pd
import matplotlib.pyplot as plt

bond_df = pd.read_csv("../data/raw/Bond_portfolio_data.csv")
```

```python
sector_mv = bond_df.groupby("Sector")["MarketValue_INR"].sum()

sector_mv.sort_values().plot(kind="barh")

plt.title("Portfolio Exposure by Sector")
plt.xlabel("Market Value (INR)")
plt.show()
```

```python
bond_df.groupby("BondType")["MarketValue_INR"] \
       .sum() \
       .plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")
plt.title("Portfolio Allocation by Bond Type")
plt.show()
```

```python
bond_df["YearsToMaturity"].hist(bins=20)

plt.title("Maturity Distribution")
plt.xlabel("Years To Maturity")
plt.ylabel("Count")
plt.show()
```

```python
bond_df["ModifiedDuration"].hist(bins=20)

plt.title("Duration Distribution")
plt.xlabel("Modified Duration")
plt.ylabel("Count")
plt.show()
```

```python
top10 = bond_df.nlargest(
    10,
    "MarketValue_INR"
)

top10[
    [
        "Issuer",
        "MarketValue_INR",
        "CreditRating"
    ]
]
```

```python
summary = {
    "Total Market Value":
        bond_df["MarketValue_INR"].sum(),

    "Average Yield":
        bond_df["YieldToMaturity"].mean(),

    "Average Duration":
        bond_df["ModifiedDuration"].mean(),

    "Average Convexity":
        bond_df["Convexity"].mean()
}

summary
```
