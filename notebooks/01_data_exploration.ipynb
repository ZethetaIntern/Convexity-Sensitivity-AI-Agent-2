```python
import pandas as pd
import matplotlib.pyplot as plt

bond_df = pd.read_csv(
    "../data/raw/Bond_portfolio_data.csv"
)

bond_df.groupby("CreditRating")["MarketValue_INR"] \
       .sum() \
       .sort_values() \
       .plot(kind="bar")

plt.title("Market Value by Credit Rating")
plt.ylabel("Market Value INR")
plt.show()
```
