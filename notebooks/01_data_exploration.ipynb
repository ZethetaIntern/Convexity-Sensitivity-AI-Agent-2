```python
import pandas as pd

bond_df = pd.read_csv("../data/raw/Bond_portfolio_data.csv")

print(bond_df.shape)
print(bond_df.head())
```

```python
import matplotlib.pyplot as plt

bond_df.columns = bond_df.columns.str.strip()

rating_mv = (
    bond_df.groupby("CreditRating")["MarketValue_INR"]
           .sum()
           .sort_values()
)

rating_mv.plot(kind="bar")

plt.title("Market Value by Credit Rating")
plt.xlabel("Credit Rating")
plt.ylabel("Market Value (INR)")
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(
    bond_df["ModifiedDuration"],
    bond_df["Convexity"]
)

plt.title("Duration vs Convexity")
plt.xlabel("Modified Duration")
plt.ylabel("Convexity")

plt.show()
```
