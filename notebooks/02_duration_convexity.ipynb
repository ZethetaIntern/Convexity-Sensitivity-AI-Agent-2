```python
import matplotlib.pyplot as plt

latest_curve = yield_df[
    yield_df["CurveDate"] == yield_df["CurveDate"].max()
]

plt.figure(figsize=(8,5))
plt.plot(
    latest_curve["Tenor_Years"],
    latest_curve["Yield"],
    marker="o"
)

plt.title("Yield Curve")
plt.xlabel("Tenor (Years)")
plt.ylabel("Yield")
plt.grid(True)

plt.show()
```
