```python
import pandas as pd
import matplotlib.pyplot as plt

mc_df = pd.read_csv("../data/raw/Monte_carlo_scenarios.csv")

print(mc_df.shape)
print(mc_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

mc_df["PnL_Total_INR"].hist(bins=40)

plt.title("Monte Carlo P&L Distribution")
plt.xlabel("PnL (INR)")
plt.ylabel("Frequency")

plt.show()
```

```python
var95 = mc_df["PnL_Total_INR"].quantile(0.05)

print(f"95% VaR = ₹{abs(var95):,.2f}")
```
